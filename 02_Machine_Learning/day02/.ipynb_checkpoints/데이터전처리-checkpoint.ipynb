### 조건에 따른 선택적인 값을 추가하기

#### 조건이 참인경우 a 아니면 b

```python
#국어점수가 70이상이면 합격 아니면 불합격을 컬럼에 넣ㅈ기
add_col['국어결과'] = numpy.where(add_col['국어']>=70, '합격', '불합격')
add_col
```

```python
#영어점수를 90이상이면 a 학점 80이상 b 70이상 c 학점 나머지 f ,만약에 빈값은 f학점
conditions = [(add_col['영어']>=90),
             (add_col['영어']>=80),
             (add_col['영어']>=70),
             (add_col['영어']<70),
             (numpy.isnan(add_col['영어']) == True)]
```

```python
#conditions 리스트 순서에 맞게, 학점부여 
grade=['A', 'B', 'C', 'F', 'F']
```

```python
add_col['영어학점'] = numpy.select(conditions, grade)
add_col
```

#### 샘플데이터만들기

### 2)열 단위 병합
#### 한번에 두개의 데이터프레임만 병합가능함

#### a)인덱스가 지정하지 않는 샘플데이터 프레임 만들기

```python
df_left = DataFrame({'고객번호':[1001,1002,1003,1004], '이름':['철수','영희','민철', '미영']})
df_left
```

```python
df_right = DataFrame({'고객번호':[1001,1002,1003,1005], '금액':[10000,20000,15000,5000]})
df_right
```

#### b) 두 데이터 프레임의 공통 컬럼을 기준으로 병합하기
양쪽 데이터 프레임에 모두 존재하는 데이터만 보여준다.
>Database의 inner join 과 같은 개념

```python
# 병합시 제외되는 항목들 
# -df_left의 고객번호 1004번 데이터는 df_right의 고객번호 컬럼과 겹치지 않는다.
# -df_right의 고객번호 1005번 데이터는 df_left의 고객번호 컬럼과 겹치지 않는다.
merge(df_left, df_right)
```

### c)왼쪽 데이터 프레임 기준으로 오른쪽 프레임을 병합하기

`how=left` 파라미터는 왼쪽 데이터프레임의 모든 데이터를 보여준다.
> 데이터베이스의 leftOuter join과 같은개념


```python
merge(df_left, df_right, how='left')
```

#### d) 오른쪽 데이터프레임 기준으로 왼쪽 프레임을 병합하기
`how=right` 파라미터는 왼쪽 데이터프레임의 모든 데이터를 보여준다.
> 데이터베이스의 leftOuter join과 같은개념


```python
merge(df_left, df_right, how='right')
```

#### e) 양쪽 데이터프레임 기준으로 왼쪽 프레임을 병합하기
`how=outer` 파라미터는 데이터프레임의 모든 데이터를 보여준다.
> 데이터베이스의 full Outer join과 같은개념

```python
merge(df_left, df_right, how='outer')
```

#### f)중복되는 데이터가 존재하는 경우 열단위 병합

```python
df_first = DataFrame({'아이디':['hello','world','python','hello'], '결제금액':[1400, 1300, 1500, 1300]})
df_first
```

```python
df_second = DataFrame({'아이디':['hello','python','python','hello'], '결제금액':[300, 500, 100, 200]})
df_second
```

```python
#아이디 명을 기준으로 병합한다.
merge(df_first,df_second, on='아이디')
```
